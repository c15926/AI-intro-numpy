```python
import numpy as np
```

상기 셀에서는 파이썬의 "numpy("넘파이"라고 발음)" 라이브러리를 "np"라는 이름으로 불러온다. Numpy 라이브러리는 기계학습 관련 코드에서 거의 항상/반드시 사용되는 라이브러리이므로 기계학습 관련 코딩을 할 때는 제일 먼저 numpy 라이브러리를 불러오도록 하자.

```python
def print_obj(obj, name):
    print("%s:\n%s\n" % (name, obj))
    
def check_each(a, b):
    return (a == b).astype('bool')

def check_mean(a, b):
    return np.mean(a == b).astype('bool')
```

위의 코드에서는 실습 과정에서 계속 사용될 함수를 정의하였다. "print_obj"는 출력하려는 객체(object)의 이름과 객체가 실제로 저장하고 있는 내용을 출력하는 역할을 한다. check_each와 check_mean은 실제 사용될 때 설명하도록 하자.

### 인덱스 (Indexing) & 슬라이싱 (Slicing)

```python
# 벡터의 인덱스와 슬라이싱 (Indexing and slicing a vector)
a = np.arange(10)
print_obj(a, "a")

print_obj(a[0], "a[0]")
print_obj(a[1], "a[1]")
print_obj(a[-1], "a[-1]")
print_obj(a[-3], "a[-3]")

print_obj(a[0:10], "a[0:10]")
print_obj(a[0:], "a[0:]")
print_obj(a[:10], "a[:10]")
print_obj(a[:], "a[:]")

print_obj(a[7:], "a[7:]")
print_obj(a[:5], "a[:5]")
print_obj(a[2:5], "a[2:5]")

# 퀴즈 3: a[-4:]는 무엇인가?
# 퀴즈 4: a[:-8]는 무엇인가?

print_obj(a[0:10:2], "a[0:10:2]")
print_obj(a[0:10:3], "a[0:10:3]")
print_obj(a[2:6:3], "a[2:6:3]")

print_obj(a[::-1], "a[::-1]")
print_obj(a[8:5:-1], "a[8:5:-1]")
print_obj(a[8:5], "a[8:5]")
```

행렬의 인덱싱을 이해하기 위해서는 인덱스 번호와 저장 순서의 차이를 이해하여야 한다. 예를 들어 [a, b, c, d]의 4개의 값이 들어있는 리스트의 첫번째 저장요소는 a, 두 번째 저장요소는 b, 세번째 저장요소는 c이다. 그런데 인덱스라는 개념을 사용할 때때는 첫번째 저장요소의 인덱스는 0, 두번째 저장요소의 인덱스는 1, i번째 저장요소의 인덱스는 i-1이다. 

**슬라이싱**

":"연산자를 이용한 인덱싱을 슬라이싱이라고 한다. 그런데 a[2:5]의 실행 결과를 살펴보자. 실제 반환된 값의 인덱스와 2:5는 어떤 관계에 있는가? a[2:5]를 실행하면 [2, 3, 4]가 반환되는데 반환된 2는 a의 세번째 값이므로 인덱스는 2(세번째 값), 반한된 3은 인덱스가 3(네번째 값), 반환된 4는 인덱스가 4(다섯번째 값)에 해당하는 값들이다. 그렇다면 "a[2:5]"라는 표현은 어떤 명령을 내리고 있다고 생각되는가? 2:5는 반환할 값들의 시작인덱스(2)와 종료인덱스(4)를 표현해야 하므로, 슬라이싱에서 **":"의 앞의 값이 시작인덱스, ":"뒤의 값-1이 종료 인덱스를 의미**한다고 정리할 수 있다.

슬라이싱 기호의 시작 인덱스와 종료 인덱스는 생략이 가능하다. 


1.  ":"기호 앞에 시작 인덱스를 생략하면 자동으로 시작 인덱스인 0으로 간주
2.   ":" 기호 뒤에 종료 인덱스를 생략하면 자동으로 맨 마지막 인덱스로 간주
3.   ":" 기호 앞 뒤를 모두 생략하면 시작부터 끝

**"-k"의 효과**

"a[-3]"의 출력 결과는 어떻게 해석할 수 있을까? a[-3]은 7을 출력하는데 7은 8번째 숫자 혹은 인덱스 7에 해당하는 값이다. 따라서 인덱싱에서 "-k"는 "원소의 갯수 - k"의 인덱스에 해당하는 값을 나타낸다고 해석할 수 있다. "-k"가 ":"기기호의 뒤에서 사용될 때는 "원소의 개수 -k - 1"의 인덱스를 표시하는 것에 주의의

**간격의 부여**

파이썬 슬라이싱에서 ":"기호가 2개 사용되는 경우는 인덱스에 간격을 부여하는 경우이다. "a[0:10:2]"와 "a[2:6:3]"의 동작 결과를 통해 간격의 동작을 추정해보자. 결과적으로는 "시작인덱스:종료값:간격"으로 명령이 전달됨을 알 수 있다.간격을 부여할 때는 "음의 간격"을 부여할 수도 있다.**음의 간격을 부여할 때는** (1) "시작인덱스:종료값:**음수**"의 형태로 표시됨에 주의해야 하며 (2) "종료값"과 반환된 값의 인덱스사이의 관계에 주의하여야 한다. 어떤 관계가 존재하는가? 종료값 +1이 실제 반환되는 값의 인덱스이다.

### 수학적 연산 (Math Operations)

```python
# 기본 연산(Basic operations)
a = np.arange(6).reshape((3, 2))
b = np.ones((3, 2))
print_obj(a, "a")
print_obj(b, "b")

# +, -, *, /
print_obj(a+b, "a+b")
print_obj(a-b, "a-b")
print_obj(a*b, "a*b")  # Hadamard product라고 함
print_obj(a/b, "a/b")
```

행렬의 곱은 NXM의 행렬과 MXK의 행렬을 곱하면 NXK의 행렬이 산출되다고 학습하였을 것이다. 그렇다면 상기 예제의 코드는 어떤 곱셈을 하고 있을까? 상기 코드에서는 3X2 행렬과 3X2 행렬을 곱하는데, 이것은 2장에서 배운 행렬곱과 맞지 않는다. 실행 결과를 비교하면 금방 알 수 있지만 "*"기호와 "/"기호를 사용하면 동일한 크기의 행렬에 대하여 A[i,j]와 B[i,j]에 대하여 원소별로 곱하기를 하거나 나누기를 한 결과를 얻게 된다(원소별로 곱하는 이런 연산을 Hadamard product라고 한다다).

2장에서 배운 행렬의 곱(NXM의 행렬과 MXK의 행렬을 곱하면 NXK의 행렬이 산출되는 경우)은 numpy의 matmul함수를 이용하거나 "@"연산자로 표현한다. "@"연산자로 표현하는 행렬의 곱셈은 노트의 하단에 설명되어 있다.

```python
# Unary 연산 (Unary operations)
a = np.arange(6).reshape((3,2))
print_obj(a, "a")

print_obj(a.sum(), "a.sum()")
print_obj(a.sum(axis=0), "a.sum(axis=0)")
print_obj(a.sum(axis=1), "a.sum(axis=1)")

print_obj(a.mean(), "a.mean()")
print_obj(a.max(), "a.max()")
print_obj(a.min(), "a.min()")

# 퀴즈 13: a = np.arange(24).reshape((2,3,4))가 주어졌을 때, 마지막 차원의 합의 평균은 얼마인가? (what is the mean of the sum w.r.t to the last dimension?)
# Hint : 퀴즈 13은 주어진 텐서에서 원하는 값들을 인덱싱한 후 mean()함수를 적용하면 된다.
```

```python
# 벡터 내적(Vector dot product)
a = np.arange(3).astype('float')
b = np.ones(3)
print_obj(a, "a")
print_obj(b, "b")

print_obj(np.dot(a, b), "a dot b")
```

```python
# 행렬 곱 연산 (Matrix dot product, matrix multiplication)
a = np.arange(6).reshape((3, 2))
b = np.ones((2, 3))
print_obj(a, "a")
print_obj(b, "b")

print_obj(np.dot(a,b), "a dot b")
print_obj(a@b, "a @ b")
```

```python
# 텐서 곱 연산 (Tensor dot product, tensor multiplication)
a = np.arange(24).reshape((4, 3, 2))
b = np.ones((4, 2, 3))
print_obj(a, "a")
print_obj(b, "b")

print_obj(np.dot(a,b).shape, "a dot b")
print_obj((a@b).shape, "a @ b")

# 퀴즈 14: a.shape==(4,3,2) and b.shape==(2,3) 이라면, 곱셈 연산이 어떻게 되는가?
# Hint : 코드로 결과를 확인하여 보자.
```

### 형태 변형 (Shape Manipulation)

```python
# Reshapes
a = np.arange(24).reshape((2, 3, 4))
print_obj(a, "a")

b = a.reshape((6, 4))
print_obj(b, "b")

c = a.reshape((6, -1))
print_obj(c, "c")

# 퀴즈 15:  d=a.reshape((6, 4, -1)) 은 어떻게 나타나는가?
```

Python의 reshape()함수의 인자로 "-1"이 주어진 경우 최초 자료형의 원소의 갯수와 -1이 아닌 인자의 값을 이용하여 "-1"로 할당된 차원의 정확한 값이 추정된다. 설명만으로는 무슨 의미인지 불명확할 것이므로 c = a.reshape(6, -1)와 e = a.reshape(-1,6)의 출력 결과를 확인하여 보자. 

```python
# 추가적인 차원 더하기 (Adding an extra dimension_
a = np.arange(3)
print_obj(a, "a")

print_obj(a[:, None], "a[:, None]")

# 퀴즈 16: a = np.ones((3,4))를 reshape와 None을 이용하여 (3, 1, 1, 4)의 shape로 변경해보기.
```

```python
# 스택(Stack), concatenation (열 등을 이어서 연결하기)
a = np.ones((3,2))
b = np.zeros((3,2))
print_obj(a, "a")
print_obj(b, "b")

print_obj(np.vstack([a, b]), "a,b vstack")
print_obj(np.hstack([a, b]), "a,b hstack")
print_obj(np.hstack([a, b, a]), "a,b hstack")

print_obj(np.concatenate([a, b], axis=0), "a,b concat axis=0")
print_obj(np.concatenate([a, b], axis=1), "a,b concat axis=1")

# 퀴즈 17: 두 텐선의 shape 가 (4, 3, 2) and (5, 4, 2) on axis=2 일 때, 두 텐서를 concatenation 하기.
```

```python
# 전치 행렬 (Matrix transpose)
a = np.arange(6).reshape((3, 2))
print_obj(a, "a")

print_obj(a.T, "a.T")
```

```python
# 전치 텐서(Tensor transpose)
a = np.arange(24).reshape((4, 3, 2))
print_obj(a, "a")

b = np.transpose(a, [0, 2, 1])
print_obj(b, "Swap axis 1 and 2")
print_obj(b.shape, "b's shape")

c = np.transpose(a, [1, 0, 2])
print_obj(c, "Swap axis 0 and 1")
print_obj(c.shape, "c's shape")
```

### 브로드케스팅(Broadcasting)

```python
# Vector and scalar
a = np.arange(3)
b = 2.
print_obj(a, "a")

print_obj(a+b, "a+b")
print_obj(a-b, "a-b")
print_obj(a*b, "a*b")
print_obj(a/b, "a/b")
```

```python
# 행렬과 벡터 (Matrix and vector)
a = np.arange(6).reshape((3,2))
b = np.arange(2).reshape(2) + 1
print_obj(a, "a")
print_obj(b, "b")

print_obj(a+b, "a+b")
```

```python
# 텐서와 행렬 (Tensor and matrix)
a = np.arange(12).reshape((2,3,2))
b = np.arange(6).reshape((3,2))
print_obj(a, "a")
print_obj(b, "b")

print_obj(a+b, "a+b")

```
